```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("wine_data.csv", header= None, usecols=[0,1,2])
df.columns = ["class_label", "alcohol", "malic acid"]
df.head()
```

```python
sns.kdeplot(df["malic acid"])
```

```python

color_dict = {1: "red", 2: "green", 3: "blue"}
sns.scatterplot(x=df["alcohol"], y=df["malic acid"], hue=df["class_label"], palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[["alcohol", "malic acid"]].values, df["class_label"].values, random_state=42, test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
mms.fit(X_train)

X_train_scale = mms.transform(X_train)
X_test_scale = mms.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns =["alcohol", "malic acid"])
X_train = pd.DataFrame(X_test, columns =["alcohol", "malic acid"])
X_train_scale = pd.DataFrame(X_train_scale, columns =["alcohol", "malic acid"])
X_test_scale = pd.DataFrame(X_test_scale, columns =["alcohol", "malic acid"])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, figsize =(12,5))
ax1.scatter(X_train["alcohol"], X_train["malic acid"])
ax1.set_title("before scaling")

ax2.scatter(X_train_scale["alcohol"], X_train_scale["malic acid"])
ax2.set_title("after scaling")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, figsize =(14,7))
ax1.set_title("before scaling")
sns.kdeplot(X_train["alcohol"], ax= ax1)
sns.kdeplot(X_train["malic acid"], ax= ax1)

ax2.set_title("after scaling")
sns.kdeplot(X_train_scale["alcohol"], ax= ax2)
sns.kdeplot(X_train_scale["malic acid"], ax= ax2)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, figsize =(12,5))
ax1.set_title("before scaling")
sns.kdeplot(X_train["alcohol"], ax= ax1)

ax2.set_title("after scaling")
sns.kdeplot(X_train_scale["alcohol"], ax= ax2)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols =2, figsize =(14,7))
ax1.set_title("before scaling")
sns.kdeplot(X_train["malic acid"], ax= ax1)

ax2.set_title("after scaling")
sns.kdeplot(X_train_scale["malic acid"], ax= ax2)

plt.show()
```

as we can see individual graphs, there might be change in shape but performance will be better because the values are now in range. outliers might afffect the performance as we are squeezing the data, so outliers will also be squeezed.
